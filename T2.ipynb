```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
%matplotlib inline
```

```python
x = np.random.rand(200,1)*10
a = np.random.rand(1,1)*10
b = np.random.rand(1,1)*10
c = np.random.rand(1,1)*10
d = np.random.rand(1,1)*10
e = np.random.rand(1,1)*10
y = (a*(x**4)) + (b*(x**3)) + (c*(x**2)) + (d*x) + e
y = y + np.random.normal(scale=3, size=y.shape)
plt.plot(x, y, 'bo')
plt.ylabel('y')
plt.xlabel('x')
print( a , b , c , d , e)
```

```python
split = int(x.shape[0] * 0.8)
```

```python
X_train = x[:split]
X_test = x[split:]

y_train = y[:split]
y_test = y[split:]
```

```python
X_train = X_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
```

```python
model = LinearRegression(normalize=True)
model.fit(X_train , y_train )
print('test score:', model.score(X_test, y_test))
print('train score:', model.score(X_train, y_train))
h_train = model.predict(X_train)
h_test = model.predict(X_test)
MSE_train = np.mean((y_train - h_train) ** 2) / 2
MSE_test = np.mean((y_test - h_test) ** 2) / 2
print('test MSE:', MSE_test)
print('train MSE:', MSE_train)
```

```python
plt.plot(X_train , y_train, 'bo',label = 'Actual Train Data')
plt.plot(X_train, h_train, 'r--',label ='Predict_Train')
plt.plot(X_test , y_test, 'co',label = 'Actual Test Data')
plt.plot(X_test, h_test, 'c--',label ='Predict_Test')
plt.ylabel('y')
plt.xlabel('x')
plt.legend()
plt.title(label = 'The power of polynomials = 1')
plt.show() 
```

```python
from sklearn.model_selection import learning_curve

model = LinearRegression(normalize=True)

train_sizes, train_score, test_score = learning_curve(model, X_train, y_train, cv=5,
                                                      train_sizes=np.arange(0.1, 1.001, 0.01))
```

```python
plt.plot(train_sizes, train_score.mean(axis=1), 'go-', label='train')
plt.plot(train_sizes, test_score.mean(axis=1), 'bo-', label='test')
plt.legend()
plt.title(label = 'The power of polynomials = 1')
plt.ylabel('acc')
plt.xlabel('Training set size')
plt.grid()
```
