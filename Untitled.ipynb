```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
%matplotlib inline
```

```python
x = np.random.rand(5,1)*10
xx = np.random.rand(25,1)*10
xxx = np.random.rand(100,1)*10
xxxx = np.random.rand(1000,1)*10
m = np.random.rand(1,1)*10
b = np.random.rand(1,1)*10
y = m*x + b
c = [x , xx , xxx , xxxx]
y = []
for i in c:
    y.append((m*i + b)+ np.random.normal(scale=3, size=i.shape))
print(m,b)
```

```python
x = x.reshape(-1, 1)
y[0] = y[0].reshape(-1, 1)
xx = xx.reshape(-1, 1)
y[1] = y[1].reshape(-1, 1)
xxx = xxx.reshape(-1, 1)
y[2] = y[2].reshape(-1, 1)
xxxx = xxxx.reshape(-1, 1)
y[3] = y[3].reshape(-1, 1)
d = [(x , y[0]),(xx,y[1]),(xxx,y[2]),(xxxx,y[3])]
```

```python
model = LinearRegression(normalize=True)
models=[]
scores_1 =[]
MSE = []
h = []
for i in d:
    models.append(model.fit(i[0], i[1]))
    scores_1.append(model.score(i[0], i[1]))
    hh = model.predict(i[0])
    h.append(hh)
    MSE.append(np.mean((i[1] - hh) ** 2) / 2)
    print(model.coef_,model.intercept_)
print(scores_1)
print(MSE)  
```

```python
d = [(x , y[0] , h[0] , 'n_sample = 5'),(xx , y[1] , h[1] , 'n_sample = 25'),(xxx ,y[2] ,h[2],'n_sample = 100' ),(xxxx ,y[3] ,h[3],'n_sample = 1000')]
for i in d:
    plt.plot(i[0], i[1], 'bo',label = 'Actual Data')
    plt.plot(i[0], i[2], 'r--',label ='Predict')
    plt.ylabel('y')
    plt.xlabel('x')
    plt.legend()
    plt.title(label = i[3])
    plt.show()
   
```

```python
e = [x , xx , xxx , xxxx]
poly_4 = []
for i in e :
    poly = PolynomialFeatures(4)
    poly_4.append(poly.fit_transform(i))

f = [(poly_4[0] , y[0]),(poly_4[1],y[1]),(poly_4[2],y[2]),(poly_4[3],y[3])]
models=[]
scores_2 =[]
MSE = []
h = []
for i in f:
    models.append(model.fit(i[0], i[1]))
    scores_2.append(model.score(i[0], i[1]))
    hh = model.predict(i[0])
    h.append(hh)
    MSE.append(np.mean((i[1] - hh) ** 2) / 2)
    print(model.coef_,model.intercept_)
print(scores_2)
print(MSE) 

```

```python
d = [(x , y[0] , h[0] , 'n_sample = 5'),(xx , y[1] , h[1] , 'n_sample = 25'),(xxx ,y[2] ,h[2],'n_sample = 100' ),(xxxx ,y[3] ,h[3],'n_sample = 1000')]
for i in d:
    plt.plot(i[0][: , 0], i[1], 'bo',label = 'Actual Data')
    plt.plot(i[0][: , 0], i[2], 'r--',label ='Predict')
    plt.ylabel('y')
    plt.xlabel('x')
    plt.legend()
    plt.title(label = i[3])
    plt.show()
   
```
